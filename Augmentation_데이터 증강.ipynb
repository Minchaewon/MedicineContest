```python
from PIL import Image
import numpy as np
import glob
import os

# 이미지 파일 경로 정의
image_files = glob.glob('/content/drive/MyDrive/식약처 공모전/가경/*/*') # 경로 수정

# 이미지를 저장할 디렉토리 생성
output_dir = '/content/drive/MyDrive/식약처 공모전/Rotated_Images'
os.makedirs(output_dir, exist_ok=True)

for idx, file in enumerate(image_files):
    # 이미지 열기
    image = Image.open(file)

    # 이미지를 numpy 배열로 변환
    image_array = np.array(image)

    # 각도를 10씩 증가시키며 회전하여 이미지 생성 및 저장
    for angle in range(0, 360, 10):
        # 이미지를 시계 방향으로 angle도 회전
        rotated_image = image.rotate(angle)

        # 회전된 이미지 파일 경로 정의
        file_name = os.path.basename(file)
        file_name_without_extension = os.path.splitext(file_name)[0]
        output_file = os.path.join(output_dir, f'{file_name_without_extension}_{angle//10}.jpg')

        # 회전된 이미지를 저장 (파일 형식을 PNG로 지정)
        rotated_image.save(output_file, format='PNG')

        # 이미지 표시
        # display(rotated_image)
    print(f'{idx+1}/{len(image_files)} - {file_name_without_extension} augmentation is done.')
```
